**Numerical Operations with NumPy**

```python
import numpy as np

# Creating an array
arr = np.array([10, 20, 30, 40])


print(arr * 2)
print(np.sqrt(arr))

# Statistical functions
print(f"Mean: {np.mean(arr)}")
print(f"Max: {np.max(arr)}")
```

**Data Manipulation with Pandas**

```python
import pandas as pd


data = {
    'Student': ['Zaid', 'Riyan', 'Tahar', 'Saima'],
    'Score': [92, 85, 88, 95],
    'Project_Completed': [True, True, False, True]
}
df = pd.DataFrame(data)

top_students = df[df['Score'] > 90]


df['Grade'] = ['A', 'B', 'B+', 'A+']

print(df)
```

**Basic Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Simple Bar Chart
plt.figure(figsize=(8, 4))
sns.barplot(x='Student', y='Score', data=df, palette='viridis')
plt.title('Student Performance')
plt.show()
```
