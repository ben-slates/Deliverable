# NumPy Operations with Stock Price Data

This notebook demonstrates essential NumPy operations using real-world stock price data for Apple Inc. (AAPL) fetched from Yahoo Finance. We'll cover array creation, indexing, reshaping, mathematical operations, statistical functions, linear algebra, and performance comparisons.

## 1. Setup and Data Loading

First, we'll import the necessary libraries and load the stock data.

```python
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv('AAPL_stock_data.csv', index_col='Date', parse_dates=True)

# Extract the 'Close' prices as a 1D NumPy array
close_prices = df['Close'].values
print(f"Close Prices (1D Array): {close_prices.shape}")
print(close_prices[:5])
```

## 2. Array Creation and Manipulation (1D, 2D, 3D)

We'll create 1D, 2D, and 3D arrays from our stock data.

```python
# 1D Array (Already created above)
a_1d = close_prices

# 2D Array: Let's create a 2D array where each row represents a week of trading (5 days)
# We'll trim the data to be a multiple of 5
num_weeks = len(a_1d) // 5
a_2d = a_1d[:num_weeks * 5].reshape(num_weeks, 5)
print(f"2D Array (Weeks x Days): {a_2d.shape}")

# 3D Array: Let's create a 3D array where each 'layer' represents a month (4 weeks per month)
num_months = num_weeks // 4
a_3d = a_2d[:num_months * 4].reshape(num_months, 4, 5)
print(f"3D Array (Months x Weeks x Days): {a_3d.shape}")
```

## 3. Array Indexing, Slicing, and Reshaping

We'll explore how to access and modify parts of our arrays.

```python
# Indexing: Accessing the price on the first day of the second week
print(f"Price on day 1 of week 2: {a_2d[1, 0]}")

# Slicing: Getting all prices for the first 4 weeks
first_4_weeks = a_2d[:4, :]
print(f"First 4 weeks shape: {first_4_weeks.shape}")

# Reshaping: Flattening the 3D array back to 1D
flattened_3d = a_3d.reshape(-1)
print(f"Flattened 3D array shape: {flattened_3d.shape}")
```

## 4. Mathematical Operations (Element-wise, Broadcasting)

We'll perform operations on our stock data.

```python
# Element-wise: Calculating daily returns (log returns)
log_returns = np.log(a_1d[1:] / a_1d[:-1])
print(f"First 5 log returns: {log_returns[:5]}")

# Broadcasting: Normalizing the prices by the mean price
mean_price = np.mean(a_1d)
normalized_prices = a_1d / mean_price
print(f"First 5 normalized prices: {normalized_prices[:5]}")

# Broadcasting with 2D array: Subtracting the weekly mean from each day's price
weekly_means = a_2d.mean(axis=1, keepdims=True)
centered_weekly_prices = a_2d - weekly_means
print(f"Centered weekly prices shape: {centered_weekly_prices.shape}")
```

## 5. Statistical Functions

We'll calculate some basic statistics for our stock data.

```python
print(f"Mean Price: {np.mean(a_1d):.2f}")
print(f"Median Price: {np.median(a_1d):.2f}")
print(f"Standard Deviation: {np.std(a_1d):.2f}")
print(f"25th Percentile: {np.percentile(a_1d, 25):.2f}")
print(f"75th Percentile: {np.percentile(a_1d, 75):.2f}")
```

## 6. Linear Algebra Operations

We'll demonstrate dot products and matrix multiplication.

```python
# Dot Product: Weighted sum of prices (e.g., for a portfolio)
weights = np.linspace(0.1, 1.0, 5) / np.sum(np.linspace(0.1, 1.0, 5))
weighted_weekly_prices = np.dot(a_2d, weights)
print(f"Weighted weekly prices (first 5): {weighted_weekly_prices[:5]}")

# Matrix Multiplication: Projecting the weekly prices onto a new basis
projection_matrix = np.random.rand(5, 3)
projected_prices = np.matmul(a_2d, projection_matrix)
print(f"Projected prices shape: {projected_prices.shape}")
```

## 7. Random Number Generation and Sampling

We'll generate random data and sample from our existing data.

```python
# Generating random normal data (simulating stock returns)
simulated_returns = np.random.normal(loc=0.0005, scale=0.01, size=1000)
print(f"Simulated returns mean: {np.mean(simulated_returns):.5f}")

# Sampling from our existing log returns
sampled_returns = np.random.choice(log_returns, size=100, replace=True)
print(f"Sampled returns mean: {np.mean(sampled_returns):.5f}")
```

## 8. Performance Comparison: NumPy vs. Python Lists

We'll compare the performance of NumPy operations with standard Python lists.

```python
size = 1000000
list_a = list(range(size))
list_b = list(range(size))
array_a = np.arange(size)
array_b = np.arange(size)

# Python List Addition
start_time = time.time()
list_sum = [a + b for a, b in zip(list_a, list_b)]
python_time = time.time() - start_time

# NumPy Array Addition
start_time = time.time()
array_sum = array_a + array_b
numpy_time = time.time() - start_time

print(f"Python List Addition Time: {python_time:.5f} seconds")
print(f"NumPy Array Addition Time: {numpy_time:.5f} seconds")
print(f"NumPy is {python_time / numpy_time:.2f}x faster than Python lists for this operation.")
```
