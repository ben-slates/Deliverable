## 5. Data Selection & Filtering

```python
# Select a single column
house_age = df['X2 house age']

# Select multiple columns
subset = df[['X2 house age', 'X3 distance to the nearest MRT station', 'Y house price of unit area']]

# Filter rows using Boolean indexing (e.g., Houses older than 20 years)
older_houses = df[df['X2 house age'] > 20]

# Apply multiple conditions (e.g., Price > 40 and distance to MRT < 500)
premium_near_mrt = df[(df['Y house price of unit area'] > 40) & (df['X3 distance to the nearest MRT station'] < 500)]

print("Filtering complete.")
display(premium_near_mrt.head())
```

## 6. Sorting Data


```python
# Sort by a single column (e.g., House price ascending)
df_sorted_price = df.sort_values(by='Y house price of unit area')

# Sort by multiple columns (e.g., Age ascending, then Price descending)
df_multi_sorted = df.sort_values(by=['X2 house age', 'Y house price of unit area'], ascending=[True, False])

print("Sorted by Age and Price:")
display(df_multi_sorted.head())
```

## 7. Handling Missing Data

```python
# Identify missing values
print("Missing values per column:")
print(df.isnull().sum())

# Fill missing values
# 1. Fill numeric 'X2 house age' with mean if there were any missing
df['X2 house age'] = df['X2 house age'].fillna(df['X2 house age'].mean())

# 2. Fill numeric 'X3 distance to the nearest MRT station' with mean
df['X3 distance to the nearest MRT station'] = df['X3 distance to the nearest MRT station'].fillna(df['X3 distance to the nearest MRT station'].mean())

print("\nMissing values after processing:")
print(df.isnull().sum())
```

## 8. Data Transformation

```python
# Add new column: PricePerYear (Price / House Age)
# Adding a small epsilon to avoid division by zero if HouseAge is 0
df['PricePerYear'] = df['Y house price of unit area'] / (df['X2 house age'] + 0.1)

# Remove columns (e.g., 'No' column is just an index and less useful)
df = df.drop(columns=['No'])

# Rename columns to be more readable
df = df.rename(columns={
    'X1 transaction date': 'TransactionDate',
    'X2 house age': 'HouseAge',
    'X3 distance to the nearest MRT station': 'DistanceToMRT',
    'X4 number of convenience stores': 'ConvenienceStores',
    'X5 latitude': 'Latitude',
    'X6 longitude': 'Longitude',
    'Y house price of unit area': 'UnitPriceArea'
})

print("Transformation complete.")
display(df.head())
```

## 9. Save Cleaned Data


```python
# Export cleaned dataset to CSV file
df.to_csv('real_estate_cleaned.csv', index=False)

print("Cleaned dataset saved as 'real_estate_cleaned.csv'.")
```
